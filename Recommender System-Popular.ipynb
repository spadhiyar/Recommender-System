```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('/Users/Zia/Google Drive/Bootcamp/Bootcamp Notes/\
Day 6 Recommendation Systems Notes/ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))
ratings.head()
```

```python
m_cols = ['movie_id', 'title']
movies = pd.read_csv('/Users/Zia/Google Drive/Bootcamp/Bootcamp Notes/\
Day 6 Recommendation Systems Notes/ml-100k/u.item', sep='|', encoding='iso-8859-1', names=m_cols, usecols=range(2))
movies.head()
```

```python
ratings = pd.merge(movies, ratings)
#print ratings 
```

```python
ratings.head()
```

```python
ratings.title.nunique(), ratings.user_id.nunique()
```

```python
movie_grouped = ratings.groupby('title').agg({'rating': [np.size, np.sum, np.mean]})
```

```python
movie_grouped
```

```python
popular_movies = movie_grouped.sort_values(('rating','mean'), ascending=False)
```

```python
popular_movies.head()
```

```python
grouped_sum = movie_grouped['rating']['sum'].sum()
grouped_sum
```

```python
popular_movies['percentage']  = popular_movies['rating','sum'].div(grouped_sum)*100
#popular_movies['percentage']  = popular_movies['rating','sum']/(grouped_sum)*100
```

```python
popular_movies.head()
```

```python
popular_movies.sort_values(('percentage'), ascending=False)
```

```python
popular_movies = popular_movies.sort_values(('percentage'), ascending=False)
```

```python
popular_movies['Rank'] = popular_movies['percentage'].rank(ascending=False)
```
